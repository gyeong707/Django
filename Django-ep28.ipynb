```python
image_url = "https://spacenews.com/wp-content/uploads/2018/05/24359364107_152b0152ff_k-879x485.jpg"
```

```python
import requests

image_binary = requests.get(image_url).content

with open("space.jpg", "wb") as f:
    f.write(image_binary)
```

```python
from PIL import Image

space_image = Image.open("space.jpg")
```

```python
space_image
```

```python
from pilkit.processors import Thumbnail
```

```python
processor = Thumbnail(width=300, height=300)
```

```python
thumb_image = processor.process(space_image)
```

```python
thumb_image
```

```python
thumb_image.save("thumb-300X300.png")
```

```python
for quality in [100, 80, 60, 40, 20, 0]:
    
    thumb_image.save("thumb-300X300-{}.jpg".format(quality), quality=quality)
```
